## Probabilidade Simples

Qual a probabilidade de ser um time com saldo negativo de gols?

```python
times_saldo_negativo = pl_table['GD'] < 0
prob_saldo_negativo = len(pl_table[times_saldo_negativo]) / len(pl_table)
print('A probabilidade de ser um time com saldo negativo é de: ' + str(round((prob_saldo_negativo * 100),4)) + '%')
```

A partir desse calculo, na premier league, apenas 45% dos times nao tem um saldo negativo de gols, ou seja, nao sofreram mais gols do que marcaram. Logo, foram os times com as melhores defesas da temporada, terminando nas primeiras 8 posicoes da liga, com excessao do Crystal Palace que possui um saldo positivo mas terminou em 12º.

Qual a probabilidade de ser um time com média de público maior que 55.000 torcedores?

```python
maior_55k = pl_table['Attendance'] > 55000
prob_maior_55k = len(pl_table[maior_55k]) / len(pl_table)
print('A probabilidade de ser um time com média maior que 55.000 torcedores é: ' + str(round((prob_maior_55k * 100),4)) + '%')
```

## Probabilidade da União

Qual a probabilidade de um time com 50 ou mais gols marcados ou 50 ou mais gols sofridos?

```python
marcados_50 = pl_table['GF'] >= 50
sofridos_50 = pl_table['GA'] >= 50

prob_marcados_50 = len(pl_table[marcados_50]) / len(pl_table)
prob_sofridos_50 = len(pl_table[sofridos_50]) / len(pl_table)

marcados_sofridos = pl_table[marcados_50 | sofridos_50]
prob_marcados_sofridos = len(marcados_sofridos) / len(pl_table)


print('A probabilidade de ser um time com 50 ou mais gols marcados é: ' + str(round((prob_marcados_50 * 100), 4)) + '%')
print('A probabilidade de ser um time com 50 ou mais gols sofridos é: ' + str(round((prob_sofridos_50 * 100), 4)) + '%')
print('A probabilidade de ser um time com um ou outro é: ' + str(round((prob_marcados_sofridos * 100), 4)) + '%')
```

## Probabilidade da Intersecção

Qual a probabilidade de ser um time com 50 ou mais gols sofridos e ter menos de 400 tackles vencidos?

```python
pl_table_temp = pl_table['Squad GA'.split()].sort_values(by=['Squad'])
defensive_actions_temp = defensive_actions['Squad TotTklW'.split()].sort_values(by=['Squad'])

merged = pl_table_temp.merge(right=defensive_actions_temp, on=['Squad'])

sofridos_50_temp = merged['GA'] >= 50
tackles_won = merged['TotTklW'] < 400
sofridos_50_TotTklW_400 = merged[sofridos_50_temp & tackles_won]

prob_50sofridos_400tacklesW = len(sofridos_50_TotTklW_400) / len(pl_table)

print('A probabilidade de ser um time com 50 ou mais gols sofridos e ter menos de 400 tackles vencidos é: ' + str(round((prob_50sofridos_400tacklesW * 100), 4)) + '%')
```

Apesar de ter menos de 400 tackles vencidos muitos times nao necessariamente tem uma defesa ruim, como é o caso do Manchester City, Liverpool e Tottenham que todos os 3 ficaram nas 4 primeiras colocaçoes. Porem, o diferencial é que nenhum desses times sofreu mais de 50 gols, significando que apesar de terem menos de 400 tackles vencidos, possuem outras formas efetivas de se defender.

Os times que tiveram tanto 50 gols sofridos quanto menos de 400 tackles vencidos foram times com diversos problemas em sua area defesiva tanto em questoes taticas quanto em elenco e terminaram a temporada em posicoes pouco prestigiadas ou ate mesmo foram rebaixados, como, o Burnley e o Norwich City. 

Ja o Manchester United teve diversos problemas em seu elenco na area defensiva como lesoes de zagueiros importantes e a qualidade de seus outros defensores, por exemplo Harry Maguire. https://www.thetimes.co.uk/article/harry-maguire-i-didnt-play-well-last-season-but-its-not-going-to-be-plain-sailing-for-your-whole-career-r5hnbmqjq


## Probabilidade do Complemento 

Qual a probabilidade de ser um time com 50 ou mais gols sofridos ou marcados e nao ter saldo negativo?

```python
prob_50sofridos_saldo_positivo = prob_marcados_sofridos - prob_saldo_negativo

print('A probabilidade de ser um time com 50 ou mais gols sofridos e nao ter saldo negativo é: ' + str(round((prob_50sofridos_saldo_positivo * 100), 4)) + '%')
```

Um time que teve 50 ou mais gols sofridos é um time que nao performou tao bem na sua defesa, porem, ainda assim conseguiu manter seu saldo de gols positivo, ou seja, marcou mais gols do que sofreu. Com isso, percebe-se que é um time que possui um ataque bom com uma defesa ruim.

## Probabilidade da Diferença

Qual a probabilidade de ser um time com 50 ou mais gols marcados e nao ter sofrido 50 ou mais gols?

```python
apenas_50_marcados = len(pl_table[marcados_50]) - len(pl_table[(sofridos_50) & (marcados_50)])

prob_apenas_50_marcados = (apenas_50_marcados) / len(pl_table)

print('A probabilidade de ser um time com 50 ou mais gols marcados sem ter sofrido 50 ou mais gols é: ' + str(round((prob_apenas_50_marcados * 100), 4)) + '%')
```

Um time que além de ter marcado 50 ou mais gols e ele ainda sofreu menos de 50 gols é um time que, no geral, performou muito bem na temporada, sendo eles: Manchester City, Liverpool, Chelsea, Tottenham, Arsenal e Crystal Palace. O Crystal Palace sendo o único time que nao performou bem como os outros nessa lista, terminando em 12º lugar.

## Probabilidade Condicional

Qual a probabilidade de ser um time que teve 50 ou mais gols sofridos sabendo que teve menos de 400 tackles vencidos?

```python
prob_menos_400_tackles = len(merged[tackles_won]) / len(merged)

prob_condicional = prob_50sofridos_400tacklesW / prob_menos_400_tackles

print('A probabilidade de ser um time que teve 50 ou mais gols sofridos sabendo que teve menos de 400 tackles vencidos é: ' + str(round((prob_condicional * 100), 4)) + '%')
```

## Teorema de Bayes

Teorema de Bayes é uma forma de calcular a probabilidade de um evento acontecer levando em consideraçao que outro ja aconteceu. Com isso, precisa-se saber a probabilidade desse evento ja ocorrido.

Qual a probabilidade de ser um time que sofreu 50 ou mais gols tal que teve menos de 400 tackles vencidos?

```python
merged_50_GA = merged['GA'] >= 50
merged_400_TotTklW = merged['TotTklW'] < 400

prob_50sofridos_400tacklesW = len(merged[merged_50_GA | merged_400_TotTklW]) / len(merged)

prob_sofreu50_tq_menos400 = (prob_condicional * prob_sofridos_50) / prob_50sofridos_400tacklesW

print('A probabilidade de ser um time com menos de 400 tackles vencidos tal que sofreu 50 ou mais gols é: ' + str(round((prob_sofreu50_tq_menos400 * 100), 4)) + '%')
```

Aqui foi calculado a probabilidade de acontecer o evento de um time ter sofrido 50 ou mais gols onde o mesmo time venceu menos de 400 tackles.

## Intervalo de Confiança

```python
def intervalo_de_confianca(array, ic):

        media = round(np.mean(array), 2)
        dp = round(np.std(array), 2)
        n = len(array)

        if ic == 0.95:
                z = 2.0930
        elif ic == 0.90:
                z = 1.7291

        print(f'media: {media}')
        print(f'desvio padrao: {dp}')
        print(f'numero da populacao: {n}')
        print(f'P({round(media - (z * (dp / np.sqrt(n))), 2)} <= μ <= {round(media + (z * (dp / np.sqrt(n))), 2)}) = {ic * 100}%\n')

intervalo_de_confianca(passing_stats['TotCmpRt'], 0.95)
intervalo_de_confianca(passing_stats['TotCmpRt'], 0.90)
```

Aqui foi criada a funçao 'intervalo_de_confianca()' que recebe como parâmetros um array, onde vai ser passado os valores da coluna que vai ser calculado o intervalo de confiaça, e vai ser passado também o nível de confiança desejado, por exemplo, 0.95 (95%). A partir disso, a funçao calcula a média dos valores da coluna usando a funçao 'mean()', depois calcula o desvio padrao dos mesmos valores. Em seguida, é analizado o nível de confiaça que foi passado e para o caso de 0.9 (90%) ou 0.95 (95%) tem-se o valor da tabela 't-student', na qual, para o meu caso, minha populaçao é de apenas 20, ou seja, possui um grau de liberdade de 19, logo, olhando em 't-student', obtem-se os valores de '2.0930' e '1.7291' para níveis de confiança de 95% e 90% respectivamente. Para finalizar, foi aplicado a fórmula para o cálculo de intervalo de confiança para os devidos valores recebidos, esses valores sendo uma coluna com a porcentagem de passes completados por cada time.
